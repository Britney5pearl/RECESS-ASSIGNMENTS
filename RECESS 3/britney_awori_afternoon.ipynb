```python
#create a series using only data from Monday and Tuesday
mydict = {"Monday": 24, "Tuesday": 27, "Wednesday": 15, "Thursday": 30, "Friday": 22}

# Filter the dictionary for Monday and Tuesday
filtered_dict = {key: value for key, value in mydict.items() if key in ["Monday", "Tuesday"]}

# Create a Series using the filtered dictionary
series = pd.Series(filtered_dict)

# Print the resulting Series
print(series)
```
